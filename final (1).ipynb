1)	According to road conditions.

```python
budget_data = [124510,47474, 15708,1209,104,132,115,75,64]
budget_categories = ['Dry','Wet','Unknown','Ice','Snow/Slush','Other','Standing Water','Sand/Mud/Dirt','Oil']
plt.pie(budget_data, labels=budget_categories, autopct='%0.1f%%')
plt.figure(figsize=(72000, 72000))
```

2)	According to Weather conditions.

```python
budget_data = [111135, 33145, 27714,15091,907,832,569,113,56,25,5]
budget_categories = ['Clear','Raining','Overcast','Unknown','Snowing','Other','Fog/Smog/Smoke','Sleet/Hail/Freezing Rain','Blowing Sand/Dirt','Severe Crosswind','Partly Cloudy']
plt.pie(budget_data, labels=budget_categories, autopct='%0.1f%%')
plt.figure(figsize=(72000, 72000))
```
